```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np

# Sample data: replace with your actual data
data = {
    'pixels': [310, 320, 330, 340, 350, 360, 370, 380],
    'real_height': [150, 155, 160, 165, 170, 175, 180, 185],
    'system_height': [152, 157, 162, 167, 172, 177, 182, 187]
}

df = pd.DataFrame(data)

# Perform Linear Regression for real height
X = df[['pixels']]
y_real = df['real_height']
model_real = LinearRegression().fit(X, y_real)
line_real = model_real.predict(X)

# Perform Linear Regression for system height
y_system = df['system_height']
model_system = LinearRegression().fit(X, y_system)
line_system = model_system.predict(X)

# Plotting
plt.figure(figsize=(8, 6))

# Scatter plot and regression line for real height
plt.scatter(df['pixels'], df['real_height'], color='black', label='Tinggi Rata-rata (pixel) vs Tinggi Real (cm)')
plt.plot(df['pixels'], line_real, color='blue', label='Regresi Linear Tinggi Real')

# Scatter plot and regression line for system height
plt.scatter(df['pixels'], df['system_height'], color='red', label='Tinggi Rata-rata (pixel) vs Tinggi Sistem (cm)')
plt.plot(df['pixels'], line_system, color='green', label='Regresi Linear Tinggi Sistem')

plt.xlabel('Pixel')
plt.ylabel('Height (cm)')
plt.title('Grafik Regresi Linear Tinggi')
plt.legend()
plt.grid(True)
plt.show()
```
